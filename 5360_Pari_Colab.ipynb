```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
df = df.rename(columns={
    '  How many trips do you make per week?  ' : 'Trips_Per_Week',
    ' What is your average daily transport cost?  ' : 'Daily_Transport_Cost',
    ' Which transport service do you use most often?  ' : 'Transport_Service',
    '  What is your main purpose of travel?  ' : 'Travel_Purpose',
    'What problem do you usually face with online transport?  ' : 'Online_Transport_Problems',
    ' Which mode do you feel is safer?  ' : 'Safer_Mode',
    ' Overall Preference  ?' : 'Overall_Preference',
    'How much time do you spend travelling daily?  ' : 'Daily_Travel_Time',
    ' How long do you usually wait for your transport?  ' : 'Waiting_Time',
    'What is the average distance of one trip?  ' : 'Avg_Trip_Distance',
    ' How often does your trip get cancelled?  ' : 'Trip_Cancellation_Freq',
    'Rate your satisfaction with offline transport services  ' : 'Offline_Satisfaction' ,
    'Rate your satisfaction with online transport services' : 'Online_Satisfaction'
})
print(df.columns)
```

```
# What is the distribution of respondents across different age groups?

plt.figure()
sns.countplot(x='Age', data=df)
plt.xlabel("Age")
plt.ylabel("Number of Respondents")
plt.title("Age Distribution of Respondents")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
# 2.How frequently do respondents travel per week, and which trip frequency category is most common?

plt.figure()
sns.countplot(x='Trips_Per_Week', data=df, color='pink')
plt.xlabel("Trips per Week")
plt.ylabel("Count")
plt.title("Weekly Trip Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
# 3.What are the most commonly used transport services among the respondents?
df.columns = df.columns.str.strip()

# Step 2: Strip spaces from column values (important)
df['Transport_Service'] = df['Transport_Service'].str.strip()
df['Gender'] = df['Gender'].str.strip()  # if using as hue

# Step 3: Define order by counts (most used at top)
order = df['Transport_Service'].value_counts().index

# Step 4: Horizontal countplot with optional hue
plt.figure(figsize=(10,6))
sns.countplot(
    x='Transport_Service',   # horizontal bars
    data=df,
    order=order,             # order by frequency
    palette='viridis',
    hue='Gender'             # optional: split by gender
)

plt.xlabel("Number of Respondents")
plt.ylabel("Transport Service")
plt.title("Most Commonly Used Transport Services by Gender")
plt.legend(title='Gender')
plt.show()

# Optional: print exact counts
print(df['Transport_Service'].value_counts())
```

```
# 4.How are respondents distributed across different average daily transport cost ranges?

plt.figure(figsize=(7,4))
sns.countplot(
    x='Daily_Transport_Cost',
    hue='Overall_Preference',
    data=df,
    palette='viridis',
    legend=True
)
plt.xlabel("Daily Transport Cost")
plt.ylabel("Count")
plt.title("Daily Transport Cost Distribution")
plt.xticks(rotation=30)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```
# 5.What is the frequency distribution of overall preference (online vs offline transport)?

plt.figure(figsize=(7,5))
df['Overall_Preference'] = df['Overall_Preference'].str.strip()

# Define order if needed
order = ['Online', 'Offline', 'Both']  # adjust based on your data

plt.figure(figsize=(6,4))
sns.countplot(
    x='Overall_Preference',
    hue='Overall_Preference',
    data=df,
    order=order,
    palette='viridis',
    legend=True
)
plt.xlabel("Overall Preference")
plt.ylabel("Number of Respondents")
plt.title("Frequency Distribution of Overall Preference")
plt.show()

# Optional: Print value counts
print(df['Overall_Preference'].value_counts())
```

```
# 6.How does trip frequency per week vary with age group?

df['Age'] = df['Age'].str.strip()
df['Trips_Per_Week'] = df['Trips_Per_Week'].str.strip()

# Crosstab of Age vs Trips_Per_Week
ct_age_trip = pd.crosstab(df['Age'], df['Trips_Per_Week'])

# Plot vertical bar chart
ct_age_trip.plot(kind='bar', figsize=(10,6))  # vertical bars
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")
plt.title("Trip Frequency per Week by Age Group")
plt.xticks(rotation=0)  # keep age labels horizontal
plt.legend(title="Trips per Week")
plt.show()
print(df['Trips_Per_Week'].value_counts())
```

```
# 7.Is there any association between average daily transport cost and transport service used most often?

ct_cost_service = pd.crosstab(df['Daily_Transport_Cost'], df['Transport_Service'])

ct_cost_service.plot(
    kind='bar',
    figsize=(9,6),
    colormap='Set3'
)
plt.xlabel("Daily Transport Cost")
plt.ylabel("Count")
plt.title("Cost vs Transport Service Used")
plt.legend(title="Transport Service")
plt.grid(True)
plt.yticks([5, 10, 15, 20, 25, 30, 35, 40])
plt.tight_layout()
plt.show()
```

```
# 8.How does main purpose of travel relate to the preferred transport mode?

ct_purpose_pref = pd.crosstab(df['Travel_Purpose'], df['Overall_Preference'])

ct_purpose_pref.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5),
    colormap='cool'
)
plt.xlabel("Travel Purpose")
plt.ylabel("Count")
plt.title("Travel Purpose vs Transport Preference")
plt.legend(title="Preference")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
# 9.How do travel time per day, waiting time, and average trip distance jointly influence overall transport preference?

df.columns = df.columns.str.strip()
# Create Crosstabs
ct_travel_time = pd.crosstab(
    df['Daily_Travel_Time'],
    df['Overall_Preference']
)

ct_waiting_time = pd.crosstab(
    df['Waiting_Time'],
    df['Overall_Preference']
)

ct_distance = pd.crosstab(
    df['Avg_Trip_Distance'],
    df['Overall_Preference']
)


# Adding labels
ct_travel_time.index = 'Travel Time: ' + ct_travel_time.index.astype(str)
ct_waiting_time.index = 'Waiting Time: ' + ct_waiting_time.index.astype(str)
ct_distance.index = 'Distance: ' + ct_distance.index.astype(str)


# Merging all crosstabs
combined_ct = pd.concat([
    ct_travel_time,
    ct_waiting_time,
    ct_distance
])


# Ploting merged crosstab
combined_ct.plot(
    kind='bar',_
    figsize=(10,7)
)

plt.xlabel('Travel Factors (Combined)')
plt.ylabel('Number of Respondents')
plt.yticks([5, 10, 15, 20, 25, 30])
plt.title('Joint Influence of Travel Time, Waiting Time, and Distance on Transport Preference')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Overall Preference')
plt.tight_layout()
plt.show()
```

```
#10.What combined effect do transport problems faced and trip cancellation frequency have on satisfaction with online transport services?

online_df = df[df['Overall_Preference'] == 'Online']

ct = pd.crosstab(
    online_df['Online_Transport_Problems'],
    online_df['Trip_Cancellation_Freq']
)

ct.plot(kind='bar', figsize=(8,7))

plt.xlabel('Online Transport Problems')
plt.ylabel('Number of Respondents')
plt.title('Problems vs Trip Cancellation Frequency (Online Users)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Trip Cancellation Frequency', fontsize=5 )
plt.tight_layout()
plt.show()
```

```
# 11.How do age group, transport service used, and safety perception together affect overall transport preference?

ct2 = pd.crosstab(
    [df['Age'], df['Transport_Service']],
    df['Safer_Mode']
)

ct2.plot(kind='bar', figsize=(9,8))

plt.xlabel('Age & Transport Service')
plt.ylabel('Number of Respondents')
plt.title('Age, Transport Service, and Safety Perception')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Safety Perception')
plt.tight_layout()
plt.show()
```

```
#12.What travel behavior patterns can be identified for respondents who prefer online transport services?

# Filter online users
online_df = df[df['Overall_Preference'] == 'Online']

# Crosstab: Travel Purpose vs Trips Per Week
ct_behavior = pd.crosstab(
    online_df['Travel_Purpose'],
    online_df['Trips_Per_Week']
)

ct_behavior.plot(
    kind='bar',
    figsize=(8,5)
)

plt.xlabel('Travel Purpose')
plt.ylabel('Number of Respondents')
plt.title('Travel Behavior Patterns of Online Transport Users')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Trips Per Week')
plt.tight_layout()
plt.show()
```

```
#13.

# Convert satisfaction levels to numeric
satisfaction_map = {
    'Very dissatisfied': 1,
    'Dissatisfied': 2,
    'Neutral': 3,
    'Satisfied': 4,
    'Very satisfied': 5
}

df['Online_Sat_Num'] = df['Online_Satisfaction'].map(satisfaction_map)
df['Offline_Sat_Num'] = df['Offline_Satisfaction'].map(satisfaction_map)

# Average satisfaction score
df['Avg_Satisfaction'] = df[['Online_Sat_Num', 'Offline_Sat_Num']].mean(axis=1)

# Plot relationship
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Trip_Cancellation_Freq',
    y='Avg_Satisfaction',
    data=df
)

plt.title("Trip Cancellation Frequency vs Average Satisfaction")
plt.xlabel("Trip Cancellation Frequency")
plt.ylabel("Average Satisfaction Level")
plt.show()
```

```
# 14.What key factors appear to most strongly influence user satisfaction with online versus offline transport?

satisfaction_factors = df.groupby('Overall_Preference')[
    ['Waiting_Time', 'Daily_Travel_Time']
].count().reset_index()

satisfaction_melted = satisfaction_factors.melt(
    id_vars='Overall_Preference',
    var_name='Factor',
    value_name='Response_Count'
)

plt.figure(figsize=(7,4))
sns.barplot(
    x='Factor',
    y='Response_Count',
    hue='Overall_Preference',
    data=satisfaction_melted,
    palette='Set1'
)

plt.xlabel('Factors')
plt.ylabel('Number of Responses')
plt.title('Key Factors Influencing Satisfaction')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
